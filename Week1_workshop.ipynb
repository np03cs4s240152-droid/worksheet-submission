###8. Create a 2d array with 1 on border and 0 inside.

```python
def create_bordered_array(rows, cols):
    arr = np.ones((rows, cols))
    if rows > 2 and cols > 2:
        arr[1:-1, 1:-1] = 0
    return arr
bordered_array = create_bordered_array(5, 5)
bordered_array
```

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
A = np.array([[3, 4],
              [7, 8]],dtype=float)

B = np.array([[5, 3],
              [2, 1]],dtype=float)

A_inv = np.linalg.inv(A)
AA_inv = np.dot(A, A_inv)
I = np.eye(2)

print("1. A @ A⁻¹ =\n", AA_inv)
print("Identity matrix I =\n", I)
print("Is A @ A⁻¹ ≈ I?", np.allclose(AA_inv, I))

AB = np.dot(A, B)
BA = np.dot(B, A)

print("\n2. AB =\n", AB)
print("   BA =\n", BA)
print("   AB == BA?", np.array_equal(AB, BA))

AB = np.dot(A, B)
AB_T = AB.T

BT_AT = np.dot(B.T, A.T)

print("\n3. (AB)^T =\n", AB_T)
print("   B^T A^T =\n", BT_AT)
print("   (AB)^T == B^T A^T?", np.allclose(AB_T, BT_AT))

A_sys = np.array([[ 2, -3,  1],
                  [ 1, -1,  2],
                  [ 3,  1, -1]], dtype=float)

C = np.array([-1, -3, 9], dtype=float)

A_inv_sys = np.linalg.inv(A_sys)
X = np.dot(A_inv_sys, C)

print("\nSolution using inverse method:")
print(f"x = {X[0]:.4f}, y = {X[1]:.4f}, z = {X[2]:.4f}")
```
