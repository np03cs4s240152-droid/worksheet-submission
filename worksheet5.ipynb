```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
#importing necessary libraries
```

```python
df=pd.read_csv('/content/drive/MyDrive/Concept and technologies of AI/dataset/Houseprice (1).csv')#Reads CSV file into a pandas DataFrame called df
print(df.head());#prints first 5 rows
```

```python
X = df[['HouseAge', 'HouseFloor', 'HouseArea']]#Selects the input features (predictors) ‚Äî these are the variables we use to predict house price.
y = df['HousePrice']#Selects the target variable (what we want to predict) ‚Äî house price.

print("\nüìä Features (X) and Target (y) ready.")
print("X shape:", X.shape)
print("y shape:", y.shape)#Confirms X has shape (100, 3) and y has (100,) ‚Äî required for training.

from sklearn.model_selection import train_test_split#Imports the train_test_split function, which randomly splits data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(#Splits the data so 80% is used for training (X_train, y_train) and 20% for testing (X_test, y_test).
    X, y,
    test_size=0.2,     #20% test.
    random_state=42    #ensures the same split every time (for reproducibility).
)
print(f"\nüîÄ Data split:")
print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")#Shows how many rows are in each set (e.g., 80 train, 20 test).

from sklearn.linear_model import LinearRegression #Imports the LinearRegression class from scikit-learn.
model = LinearRegression()#Creates an instance of the linear regression model (but doesn‚Äôt train it yet).
model.fit(X_train, y_train)#Trains the model on the training data.

print("\n‚úÖ Model trained successfully!")

from sklearn.metrics import mean_squared_error, r2_score#Imports metrics to evaluate model performance.

y_pred = model.predict(X_test)#Uses the trained model to predict house prices for the test set (X_test). Results stored in y_pred.

mse = mean_squared_error(y_test, y_pred)#Average squared difference between actual and predicted prices. Lower = better.
rmse = mse ** 0.5 #Square root of MSE ‚Äî in the same units as price (easier to interpret).
r2 = r2_score(y_test, y_pred) #Proportion of variance in price explained by the model (0 to 1). Higher = better.

print("\n" + "="*50)
print("üìà MODEL PERFORMANCE")
print("="*50)
print(f"R¬≤ Score (Explained Variance): {r2:.4f} ({r2*100:.2f}%)") #Shows how well the model fits.
print(f"Mean Squared Error (MSE): {mse:,.2f}") # Prints MSE with commas for readability
print(f"Root Mean Squared Error (RMSE): {rmse:,.2f}") #Prints RMSE ‚Äî average prediction error in dollars

print("\nüìê MODEL EQUATION:")
print(f"HousePrice = ({model.coef_[0]:.2f} √ó HouseAge) + "
      f"({model.coef_[1]:.2f} √ó HouseFloor) + "
      f"({model.coef_[2]:.2f} √ó HouseArea) + {model.intercept_:.2f}")#Displays the regression equation ‚Äî this is the core result!

results = pd.DataFrame({
    'Actual Price': y_test,
    'Predicted Price': y_pred,
    'Error': y_test - y_pred
})#Creates a new DataFrame comparing actual vs. predicted prices, and their difference (error).

results = results.round(2)#Rounds all numbers to 2 decimal places for clean output.
print("\nüîç Sample Predictions (First 10):")
print(results.head(10))#Shows the first 10 predictions so you can visually assess accuracy.
```
